```python
import serial
import numpy as np
from time import time
```

```python
def readserial(comport, baudrate):

    ser = serial.Serial(comport, baudrate, timeout=0.1) 
    
    while True:
        data = ser.readline().decode().strip()
        print(data)

readserial('/dev/ttyACM0', 115200, True)     
```
